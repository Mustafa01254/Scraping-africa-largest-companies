```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_Africa_by_revenue'
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
table = soup.find('table', class_='wikitable sortable')
print(table)
```

```python
table = soup.find_all('table')[1]
print(table)
```

```python
table_heads = table.find_all('th')
table_heads
```

```python
table_title = [title.text.strip() for title in table_heads]
table_title
```

```python
df = pd.DataFrame(columns = table_title)
df
```

```python
companies_data = table.find_all('td')
africa_data = [data1.text.strip() for data1 in companies_data]
africa_data
```

```python
column_data = table.find_all('tr')
column_data
```

```python
for row in column_data[1:]:
    table_data = row.find_all('td')
    africa_table_data = [data.text.strip() for data in table_data]
    
    length = len(df)
    df.loc[length] = africa_table_data
    print(df)
    
```
